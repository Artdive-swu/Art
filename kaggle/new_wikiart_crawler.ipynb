```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
kaggle = pd.read_csv('openrefine_kaggle_main(작업중).csv', index_col = 'Column')

print(kaggle.shape)
kaggle.head(5)
```

```python
kaggle_copy = kaggle.copy()
```

```python
artist_list = kaggle_copy['Artist'].unique()
```

```python
pd.DataFrame(columns=['Artist', 'Art', 'Link'])
```

```python
kaggle_copy.loc[kaggle_copy['Artist']=='Barnett Newman', 'Title']
```

```python
df = pd.DataFrame(columns=['Artist', 'Art', 'Link'])

for i in range(len(artist_list)):
    print(i, artist_list[i])
    artist = artist_list[i].lower().replace(' ', '-')
#     title_list = kaggle_copy.loc[kaggle_copy['Artist']=='Barnett Newman', 'Title']

    url = f'https://www.wikiart.org/en/{artist}/all-works/text-list'

    response = requests.get(url)  
    html = response.text

    soup = BeautifulSoup(html, 'html.parser')

    for artwork in soup.select('.painting-list-text a'):
        link = 'https://www.wikiart.org' + artwork['href']
        title = artwork.text
        
        tmp = pd.DataFrame([{'Artist': artist_list[i], 'Art': title, 'Link' : link}])
        
        df = pd.concat([df, tmp])
        
```

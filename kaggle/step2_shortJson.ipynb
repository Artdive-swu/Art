```
new_shortjson_list=[]
for i in range(len(all_short_json)):
  try:
    all_short_json[i]['Exception']

    title = kaggle_copy.loc[i, 'Title']

    if pd.isnull(title):
      new_shortjson_list.append('non-title')
      continue

    url = f'https://www.wikiart.org/en/api/2/PaintingSearch?term={title}'
    response = requests.get(url)

    json_file = json.loads(response.text)

    new_shortjson_list.append(json_file)
    print(i)

  except:
    new_shortjson_list.append(all_short_json[i])
```

```
#1.json 부분 수정
file_path = kaggle_path+'new_0.json'

with open(file_path, 'w') as outfile:
    json.dump(new_shortjson_list, outfile)

#shotjson_list
```

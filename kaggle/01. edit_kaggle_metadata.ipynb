**컬럼 Index 추가 : Kaggle_ + new_filename**

```
#filename의 .jpg를 제외한것을 ID로 만들기
Kaggle['filename_split']=Kaggle['new_filename'].str.split(".").str[0]
Kaggle['Index'] = Kaggle['filename_split'].map(lambda x : 'Kaggle_'+x)

del Kaggle['filename_split']
```

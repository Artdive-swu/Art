```
all_metadata = []
for file in os.listdir(path):
  print(file)
  xml = ET.parse(path + file)
  xtree = xml.getroot()

  idx = int(file.split('.')[0])
  
  cnt = 0
  ListRecords = xtree.find('{http://www.openarchives.org/OAI/2.0/}ListRecords')
  for record in ListRecords.findall('{http://www.openarchives.org/OAI/2.0/}record'):
    metadata_dic = {}
    metadata = record.find('{http://www.openarchives.org/OAI/2.0/}metadata')
    dc = metadata.find('{http://www.openarchives.org/OAI/2.0/oai_dc/}dc')

    metadata_dic['idx'] = idx + cnt

    tmp = []
    for identifier in dc.findall('{http://purl.org/dc/elements/1.1/}identifier'):
      tmp.append(identifier.text)
    metadata_dic['identifier'] = tmp

    tmp = []
    for format in dc.findall('{http://purl.org/dc/elements/1.1/}format'):
      tmp.append(format.text)
    metadata_dic['format'] = tmp

    metadata_dic['language'] = dc.find('{http://purl.org/dc/elements/1.1/}language').text

    metadata_dic['publisher'] = dc.find('{http://purl.org/dc/elements/1.1/}publisher').text

    metadata_dic['rights'] = dc.find('{http://purl.org/dc/elements/1.1/}rights').text

    tmp = []
    for date in dc.findall('{http://purl.org/dc/elements/1.1/}date'):
      tmp.append(date.text)
    metadata_dic['date'] = tmp

    metadata_dic['title'] = dc.find('{http://purl.org/dc/elements/1.1/}title').text

    metadata_dic['description'] = dc.find('{http://purl.org/dc/elements/1.1/}description').text

    metadata_dic['creator'] = dc.find('{http://purl.org/dc/elements/1.1/}creator').text

    metadata_dic['type'] = dc.find('{http://purl.org/dc/elements/1.1/}type').text

    tmp = []
    for subject in dc.findall('{http://purl.org/dc/elements/1.1/}subject'):
      tmp.append(subject.text)
    metadata_dic['subject'] = tmp

    all_metadata.append(metadata_dic)
    # print(metadata_dic['idx'], metadata_dic)

    cnt += 1
```

```
real_test = pd.DataFrame(all_metadata)
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
kaggle_wikilink = pd.read_csv('kaggle_wikilink_0.csv', index_col='Column')

print(kaggle_wikilink.shape)
kaggle_wikilink.head()
```

```python
for num in range(10000, 100000, 5000):
    try:
        tmp = pd.read_csv(f'kaggle_wikilink_{num}.csv', index_col='Column')
        
        print(f'{num} O {len(tmp)}')
        
        kaggle_wikilink = pd.concat([kaggle_wikilink, tmp])

    except:
        print(f'{num} X')

    # break
```

```python
artist_list = kaggle_wikilink['Artist'].unique()
```

```python
artist_list[1360:1370]
```

```python
df = pd.DataFrame(columns=['Artist', 'Art', 'Link'])

for i in range(1360, len(artist_list)):
    print(i, artist_list[i])
    if(pd.isnull(artist_list[i])):
        print('NaN')
        continue
    artist = artist_list[i].lower().replace(' ', '-')
#     title_list = kaggle_copy.loc[kaggle_copy['Artist']=='Barnett Newman', 'Title']

    url = f'https://www.wikiart.org/en/{artist}/all-works/text-list'

    response = requests.get(url)  
    html = response.text

    soup = BeautifulSoup(html, 'html.parser')

    for artwork in soup.select('.painting-list-text a'):
        link = 'https://www.wikiart.org' + artwork['href']
        title = artwork.text
        
        tmp = pd.DataFrame([{'Artist': artist_list[i], 'Art': title, 'Link' : link}])
        
        df = pd.concat([df, tmp])
        
```

```python
df.to_csv('kaggle_wikilink2.csv')
```

```python
for i in range(len(kaggle_wikilink)):
    if kaggle_wikilink.loc[i, 'tmp'] == 'no_link':
        artist = kaggle_wikilink.loc[i, 'Artist']
        title = kaggle_wikilink.loc[i, 'Title']
        
        link = list(df.loc[(df['Artist'] == artist) & (df['Art']==title), 'Link'])
        
        kaggle_wikilink.loc[i, 'Wiki_Link'] = link
```

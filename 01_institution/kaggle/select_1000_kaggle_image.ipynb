```
def extract_year(date):
  if pd.isnull(date):
    return date

  if '.' in date:
    return date.split('.')[1]
  elif ' ' in date:
    return date.split(' ')[1]
  else:
    return date
```

```
kaggle['date_year'] = kaggle['date'].apply(extract_year)
```

```
style_list_count = {}
index = []

for i in range(len(kaggle)):
  try:
    if style_list_count[kaggle.loc[i, 'style']] < 10:
      style_list_count[kaggle.loc[i, 'style']] += 1
      index.append(i)
  except:
    style_list_count[kaggle.loc[i, 'style']] = 0
    index.append(i)
```

```
artist_list_count = {}
index = []

for i in range(len(kaggle)):
  try:
    if artist_list_count[kaggle.loc[i, 'artist']] < 0:
      artist_list_count[kaggle.loc[i, 'artist']] += 1
      index.append(i)
  except:
    artist_list_count[kaggle.loc[i, 'artist']] = 0
    index.append(i)
```

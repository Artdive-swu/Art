```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
kaggle = pd.read_csv('openrefine_kaggle_main(작업중).csv', index_col = 'Column')

print(kaggle.shape)
kaggle.head(5)
```

```python
kaggle_copy = kaggle.copy()
```

```python
artist_list = kaggle_copy['Artist'].unique()
```

```python
pd.DataFrame(columns=['Artist', 'Art', 'Link'])
```

```python
kaggle_copy.loc[kaggle_copy['Artist']=='Barnett Newman', 'Title']
```

```python
requests.get('https://www.wikiart.org/en/kiri-nichol/all-works/text-list')  
```

```python
df = pd.DataFrame(columns=['Artist', 'Art', 'Link'])

for i in range(len(artist_list)):
    print(i, artist_list[i])
    artist = artist_list[i].lower().replace(' ', '-')
#     title_list = kaggle_copy.loc[kaggle_copy['Artist']=='Barnett Newman', 'Title']

    url = f'https://www.wikiart.org/en/{artist}/all-works/text-list'

    response = requests.get(url)  
    html = response.text

    soup = BeautifulSoup(html, 'html.parser')

    for artwork in soup.select('.painting-list-text a'):
        link = 'https://www.wikiart.org' + artwork['href']
        title = artwork.text
        
        tmp = pd.DataFrame([{'Artist': artist_list[i], 'Art': title, 'Link' : link}])
        
        df = pd.concat([df, tmp])
        
```

```python
df.pivot_table(index='Artist', values='Link', aggfunc =len)
```

```python
list(df.loc[(df['Artist'] == 'Hiroshige') & (df['Art']=='Small Bird on a Branch of Kaidozakura'), 'Link'])
```

```python
def find_link(row):
    artist = row['Artist']
    title = row['Title']
#     print(artist, title)
    link = list(df.loc[(df['Artist'] == artist) & (df['Art']==title), 'Link'])
    
    return link

kaggle_copy['Wiki_Link'] = kaggle_copy.apply(find_link, axis = 1)
```

```python
new_kaggle = kaggle_copy[['Art_ID', 'Title', 'Artist', 'Wiki_Link']]
```

```python
new_kaggle.to_csv('kaggle_wikilink.csv')
```

```python
for i in range(3780, 10000):
    print(i, end=' ')
    links = new_kaggle.loc[i, 'Wiki_Link']
    
    if len(links) == 0:
        new_kaggle.loc[i, 'tmp'] = 'no_link'
    elif len(links) > 1:
        new_kaggle.loc[i, 'tmp'] = 'check'
    else:
        try:
            url = links[0]

            print(url)

            response = requests.get(url)  
            html = response.text

            soup = BeautifulSoup(html, 'html.parser')

            article = soup.select_one('article')
            for info in article.select('.dictionary-values'):
                info = info.text
                key, value = info.split(':')

                new_kaggle.loc[i, f'{key.strip()}'] = value.strip()
        except:
            new_kaggle.loc[i, 'tmp'] = 'error'
            print('error')
```

```python
new_kaggle[0:10000].to_csv('./kaggle_wikilink_0.csv')
```

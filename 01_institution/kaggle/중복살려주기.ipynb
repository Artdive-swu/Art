```python
import os
import pandas as pd
from ast import literal_eval
import requests
from bs4 import BeautifulSoup
```

```python
kaggle_wikilink = pd.read_csv('./dataset/kaggle_wikilink_all_finish.csv', \
                     index_col ='Column', converters={"Wiki_Link": literal_eval})
```

```python
check_kaggle = kaggle_wikilink[kaggle_wikilink['tmp'] == 'check']
```

```python
print(check_kaggle.shape)
check_kaggle.head(5)
```

```python
check_kaggle_unique = pd.read_csv('tmp.csv', index_col='Unnamed: 0', converters={"Wiki_Link": literal_eval})

print(check_kaggle_unique.shape)
check_kaggle_unique.head()
```

```python
check_kaggle_unique = check_kaggle_unique[['Title', 'Artist', 'New_Wiki_Link', 'img_src']]

check_kaggle_unique.head()
```

```python
new_check_kaggle = pd.merge(check_kaggle, check_kaggle_unique, how='left', on = ['Title', 'Artist'])
```

```python
new_check_kaggle = new_check_kaggle[~(new_check_kaggle['New_Wiki_Link'] == "[]")]
```

```python
new_check_kaggle = new_check_kaggle[['Art_ID', 'Title', 'Artist', 'New_Wiki_Link', 'img_src']]
```

```python
print(new_check_kaggle.shape)
new_check_kaggle.head()
```
